**Operator aritmatika**

Operator yang akan kita bahas adalah penambahan, pengurangan, perkalian, eksponensial, modulus, dan pembagian. Mari kita lihat operasi aritmatika sederhana pada bahasa pemrograman Python.

```python
# Pertamabahan
a = 2 + 3
print(a)

# Pengungaran
a = 2 - 5
print(a)

# Perkalian
a = 2 * 3
print(a)

# Pembagian
a = 2 / 5
print(a)

# Eksponensial
a = 2 // 5
print(a)
```

**Operator Pembanding**

Operator pembanding
Contoh operator pembanding adalah sama dengan, lebih besar dari, serta kurang dari

```python
# Contoh operator pembanding
a = 2
b = 5

u = a > b
v = a < b
w = a == b
x = a >= b
y = a <= b
z = a != b
print(u)
print(v)
print(w)
print(x)
print(y)
print(z)


```

**Operator Logika**

operator ini "not" "and" "or"

```python
# Contoh operator logika
a = 2
b = 5
# not
not(a == 2) # output false

# and
a == 2 and b == 5 # output true

# or
a == 2 or b == 10 # output true
```
